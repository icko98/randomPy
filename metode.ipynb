```python
def ackley(x):
    sum1 = 0
    sum2 = 0
    
    for y in x:
        sum1=sum1+ y*y
        sum2=sum2+ np.cos(np.pi*2*y)
    

    ret = -20 * np.exp(-0.2 * np.sqrt(1/len(x) * sum1)) - np.exp(1/len(x) * sum2) + 20 + np.exp(1)
    return ret


x1v = np.arange(-5,5,0.01)
x2v = np.arange(-5,5,0.01)
[x1,x2] = np.meshgrid(x1v,x2v)

f=np.zeros((len(x1), len(x2)))


for i in range(0, len(x1)):
    for j in range(0, len(x2)):
        f[i,j] = ackley((x1[i,j], x2[i,j]))
        
                        
fig = plt.figure()
ax = fig.gca(projection = '3d')                        
p=ax.plot_surface(x1,x2,f)
plt.show()
print("???")
```

```python
def Griewank(x):
    n = len(x)
    sum1 = 0
    pro1 = 1
    ii = 1
    
    for y in x:
        sum1 = sum1 + ((y*y) / 4000)
        pro1 = pro1 * np.cos(y / np.sqrt(ii))
        ii=ii+1
    return sum1-pro1+1

x1v = np.arange(-5,5, 0.01)
x2v = np.arange(-5,5, 0.01)

x1,x2 = np.meshgrid(x1v,x2v)
f = np.zeros((len(x1), len(x2)))

for i in range(len(x1)):
    for y in range(len(x2)):
        f[i,y] = Griewank((x1[i,j], x2[i,j]))
fig = plt.figure()
ax = fig.gca(projection = '3d')                        
p=ax.plot_surface(x1,x2,f)
plt.show()
```
