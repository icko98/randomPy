```python
#ok = data1['opetkancer'].notna()
#ok2 = data1.loc[data1['kancer']==np.nan]
ok3 = data1['opetkancer'].dropna()
indexi = ok3.index # indeksi svih redova gde imamo 8 kolona umesto 6

data2=data1.copy() #failsafe

for x in indexi: # shift kolona ulevo
    data2.iloc[x,3]= data1.iloc[x,3]+data1.iloc[x,4]
    data2.iloc[x,4]= data1.iloc[x,5]
    data2.iloc[x,5]= data1.iloc[x,6]
    data2.iloc[x,6]= data1.iloc[x,7]
data2 = data2.drop(columns='opetkancer')
data2.iloc[27]

```

```python
ok4 = data2['kancer'].dropna()
indexi2 = ok4.index # indeksi svih redova gde imamo 7 kolona umesto 6
data3=data2.copy()

for x in indexi2: # shift kolona ulevo
    data3.iloc[x,3]= str(data3.iloc[x,3])+str(data3.iloc[x,4])
    data3.iloc[x,4]= data3.iloc[x,5]
    data3.iloc[x,5]= data3.iloc[x,6]

    
data3=data3.drop(columns='kancer')
data3=data3.drop(columns='pib')
data3=data3.drop(columns='adresa')
data3['grad']=data3['grad'].apply(lambda x: x[1:])
```

```python
kladionice = data3.loc[data3['tip']==' кладионица']
grupaklad = kladionice.groupby('grad')['broj'].count()

kladionice.loc[kladionice['grad']=='Палилула']
```

```python
grad1 = pd.read_table(r"C:\Users\Ilija\Desktop\kim\cenzus2.csv",header=0, sep=';')
grad1 = grad1.drop(columns=['IDVrPod','nVrPod', 'IDIndikator' , 'skrIndikator', 'IDTer' ,'nPol' , 'IDStatusPodatka', 'nStatusPodatka', 'IDStatusPoverljivosti', 'nStatusPoverljivosti' , 'IDJedinicaMere', 'nJedinicaMere'  ])
grad1 = grad1.rename(columns={'nTer' : 'grad'})
grad1
```

```python

merged = pd.merge(kladionice, grad1, on=['grad'])
#merged.loc[merged['grad']=='Ниш']

grupa2 = merged.groupby('grad').agg({"broj":"count", "Vrednost":"first"})
grupa2['xx'] = grupa2['broj'] * 10000 / grupa2['Vrednost']
grupa2 =grupa2.sort_values(by=['xx'],ascending=False)
grupa2[:15]
```
